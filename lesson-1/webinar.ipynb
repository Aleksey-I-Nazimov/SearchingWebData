```python
params={
    'lang': 'ru',
    'query': 'морковь'
}
headers = {
    'User-agent': "Mozilla/5.0 (Windows NT 10.0; Win64;x64)"
}
rs = rq.get(url='https://www.delivery-club.ru/search',params=params);
print (str(rs.headers)+"\n\n")
print (rs.content)
```

```python
params={
    'lang': 'ru',
    'query': 'морковь'
}
headers = {
    'User-agent': "Mozilla/5.0 (Windows NT 10.0; Win64;x64)",
    'Origin': "https://www.delivery-club.ru",
    'Referer': "https://www.delivery-club.ru/search?query=морковь&type="
}
json = {"text":"морковь","region_id":1}
rs = rq.post(url='https://www.delivery-club.ru/eats/v1/full-text-search/v1/search',params=params,json=json);
print (str(rs.json)+"\n\n")
json = rs.json();
print (rs.content)
```
