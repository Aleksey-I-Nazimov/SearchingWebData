1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
# https://docs.github.com/ru/search-github/searching-on-github/searching-for-repositories
! curl -X GET "https://api.github.com/search/issues?q=user:Aleksey-I-Nazimov&order=asc"
```

```python
import requests as rq
import json

params={
    'q': 'user:Aleksey-I-Nazimov',
    'order': 'asc'
}
url="https://api.github.com/search/repositories"

response = rq.get(url=url,params=params);
response_body = response.json();
repository_list=[];

for repository in response_body['items']:
    repository_list.append({'name':repository['name'],'url':repository['url']});

print ("Repository list: {}".format(repository_list));

with open('repositories.txt', 'w') as f:
  json.dump(repository_list, f, ensure_ascii=False)
```
